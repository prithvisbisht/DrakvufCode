```python
import csv
import os
fname="citadel2.txt"
import pandas as pd
```

```python
count=0
# process={}
process={"filename":fname.split(".")[0]}
with open(fname,"r",encoding="utf8", errors='ignore') as file:
    file.seek(0)
    for line in file:
        if line.find("!Nt")!= -1:
            count+=1
            word=line.split("!")[1]
            word=word.split(" ")[0]
            word=word[2:]
            if word in process.keys():
                process[word]+=1
            else:
                process[word]=1

print(process)
# print(sum(process.values()))
print(count)
file.close()
```

```python
with open('process.xls','a+') as my_file:
    my_file.seek(0)
    first_char = my_file.read(1)
    if not first_char:
        empty=1
    else:
        empty=0
my_file.close()
with open("process.xls","a+") as file:
    w = csv.DictWriter(file, process.keys())
    print(w)
    if empty==1:
        w.writeheader()
    w.writerow(process)
file.close()
```
