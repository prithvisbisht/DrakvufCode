```python
import csv
import os
import pandas as pd
dataset={}
```

```python
fname="a/wiwrar.txt"
count=0
process={}
with open(fname,"r",encoding="utf8", errors='ignore') as file:
    file.seek(0)
    for line in file:
        if line.find("!Nt")!= -1:
            count+=1
            word=line.split("!")[1]
            word=word.split("\n")[0]
            word=word[2:]
            if word in process.keys():
                process[word]+=1
            else:
                process[word]=1

print(process)
# print(sum(process.values()))
print(count)
file.close()
len(process)
```

```python
dataset.update({fname:process})
```

```python
df = pd.DataFrame(dataset).T
df
```

```python
df.to_excel('process.xls')
```
